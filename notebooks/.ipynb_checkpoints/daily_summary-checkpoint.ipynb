## Reoptimize spin polarizing

Typically if we do spin polarizing, we run an "anti_polarize" sequence where half the atoms end up in say +9/2 and the other half in -9/2, so that we may then evaporative. if we aren't evaporating, we only need one spin state after spin polarization.
On 20171130 we added new sequences "polarize_m" and "polarize_p" which try to put all of the atoms in either -9/2 or +9/2. 

on 20171130 we noticed large (~75%) difference in atom numbers between $\sigma_{+/-}$ spin polarization (sigma plus was the bad one). This was attributed to imperfect polarization of the 689 nm light during the "polarize_p" sequence so we decided to redo the spin-pol setup.
